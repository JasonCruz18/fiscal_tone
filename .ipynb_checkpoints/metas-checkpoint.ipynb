```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Rectangle, Patch
from matplotlib.ticker import FormatStrFormatter

# Datos
data = {
    'year': [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015,
             2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025],
    'y_value': [1.2, 0.8, 1.1, -0.3, -0.4, -0.6, -3.0, -4.1, -3.1, -0.1, -1.5, -2.9, -3.2, -0.9, 0.3, 1.1,
                -0.4, 0.0, -1.2, -1.3, 6.8, 0.7, -2.0, 0.4, 0.8, 0.4],
    'meta': ["Meta incumplida", "Regla suspendida", "Regla suspendida", "Meta cumplida", "Meta cumplida",
             "Meta cumplida", "Meta cumplida", "Meta cumplida", "Meta cumplida", "Meta cumplida", "Meta cumplida",
             "Meta cumplida", "Meta cumplida", "Meta cumplida", "Meta incumplida", "Meta cumplida", "Meta cumplida",
             "Meta cumplida", "Meta cumplida", "Meta cumplida", "Regla suspendida", "Regla suspendida",
             "Meta cumplida", "Meta incumplida", "Meta incumplida", "Meta incumplida"]
}

df = pd.DataFrame(data)
df['size'] = abs(df['y_value']) * 600

color_map = {"Meta cumplida": "#033280", "Meta incumplida": "#E6004C"}

fig, ax = plt.subplots(figsize=(10, 6))

# Bandas
ax.add_patch(Rectangle((2001, -6), 2, 14, color="#F5F5F5", alpha=1))
ax.add_patch(Rectangle((2020, -6), 2, 14, color="#F5F5F5", alpha=1))

# Burbujas
for meta in ["Meta cumplida", "Meta incumplida"]:
    meta_data = df[df['meta'] == meta]
    ax.scatter(meta_data['year'], meta_data['y_value'],
               s=meta_data['size'],
               c=meta_data['meta'].map(color_map),
               edgecolors='black', linewidths=1, alpha=0.85)

# Lnea horizontal en 0
ax.axhline(0, color='#292929', linestyle='-', linewidth=1)

# Estrella
ax.scatter([2015], [1.85], s=200, c="#00DFA2", alpha=0.9,
           edgecolors="black", linewidths=0.75, marker="*", zorder=5)

# Ejes
ax.set_ylim(-6, 6)
ax.set_yticks(np.arange(-4, 6, 2))
ax.set_ylabel("Ejecucin  Meta (% del PIB)", fontsize=14)
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

ax.set_xlim(1999.25, 2025.75)
xticks = np.arange(2000, 2026, 5)
xticklabels = [f"{x} (P)" if x == 2025 else str(x) for x in xticks]
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, fontsize=12)

# Leyenda
legend_handles = [
    plt.scatter([], [], s=200, c="#033280", alpha=0.85, edgecolors="black", label="Meta cumplida"),
    plt.scatter([], [], s=200, c="#E6004C", alpha=0.85, edgecolors="black", label="Meta incumplida"),
    Patch(facecolor="#F5F5F5", edgecolor="#F5F5F5", label="Regla suspendida")
]
legend = ax.legend(handles=legend_handles, loc='upper left', frameon=True, fontsize=12)
legend.get_frame().set_facecolor('white')
legend.get_frame().set_alpha(0.5)
legend.get_frame().set_edgecolor('#f5f5f5')
legend.get_frame().set_linewidth(2)

# Rejilla
ax.grid(True, color='#f5f5f5', alpha=0.75, linewidth=1, zorder=0)

# Bordes
for spine in ax.spines.values():
    spine.set_edgecolor('#292929')
    spine.set_linewidth(0.80)
    spine.set_zorder(3)

# FORZAR TICKS VISIBLES
ax.tick_params(axis='both', which='major', length=8, width=1.5, color='black', direction='out', zorder=4)
ax.minorticks_on()  # activa minor ticks

plt.tight_layout()
plt.savefig("agenda_3.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import numpy as np
from matplotlib import cm
from matplotlib.colors import Normalize

# Cargar el archivo Excel
file_path = 'mef_ministers_dataset.xlsx'  # Asegrate de cambiar esta ruta
df = pd.read_excel(file_path)

# Asegrate de que las fechas estn en formato datetime
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')  # Convertir a datetime y manejar errores
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')

# Eliminar filas con fechas nulas
df = df.dropna(subset=['start_date', 'end_date'])

# Calcular la duracin en das
df['duration_days'] = (df['end_date'] - df['start_date']).dt.days

# Ordenar por la fecha de inicio
df = df.sort_values('start_date')

# Crear la figura para el grfico
fig, ax = plt.subplots(figsize=(14, 6))

# Normalizacin de la duracin para el mapa de calor
norm = Normalize(vmin=300, vmax=900)

# Graficar las lneas de tiempo
for i, row in df.iterrows():
    color = cm.viridis(norm(row['duration_days']))  # Usar un mapa de color para la duracin
    ax.plot([row['start_date'], row['end_date']], [i, i], linewidth=5, color=color)

# Ajustar el formato de las fechas en el eje X
ax.xaxis.set_major_locator(mdates.YearLocator(2))  # Mostrar cada 2 aos
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gcf().autofmt_xdate()

# Eje Y: Ministros
ax.set_yticks(np.arange(len(df)))
ax.set_yticklabels(df['minister_name'])

# Eje Y: Etiqueta
ax.set_ylabel("Ministro", fontsize=16)

# Eje X: No tiene etiqueta
ax.set_xlabel('')

# Rejilla esttica
ax.grid(True, color='#f5f5f5', alpha=0.75, linewidth=1)

# Lnea horizontal en 0 (simulando un tono neutral en el eje Y)
ax.axhline(0, color='#292929', linestyle='-', linewidth=1)

# Personalizacin de los ticks en el eje Y y X
ax.tick_params(axis='x', labelrotation=0, labelsize=12)  # Rotar etiquetas X a 0 (horizontal)
ax.tick_params(axis='y', labelsize=12)

# Leyenda personalizada (escala de colores de duracin en das)
sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])  # Necesario para la leyenda
cbar = fig.colorbar(sm, ax=ax, orientation='horizontal', pad=0.1)
cbar.set_label('Duracin en das', fontsize=12)
cbar.set_ticks([300, 600, 900])  # Los valores que aparecen en la leyenda de la escala

# Mover la leyenda a la parte superior izquierda
ax.legend(['Das en la Oficina'], loc='upper left', frameon=True)
ax.get_legend().get_frame().set_facecolor('white')
ax.get_legend().get_frame().set_alpha(0.5)
ax.get_legend().get_frame().set_edgecolor('#f5f5f5')  # contorno sin transparencia
ax.get_legend().get_frame().set_linewidth(2)  # grosor del contorno

# Ajuste final de la figura
plt.tight_layout()

# Guardar el grfico
plt.savefig('ministers_tenure.png', dpi=300)

# Mostrar el grfico
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import numpy as np
from matplotlib import cm
from matplotlib.colors import Normalize, ListedColormap

# Cargar el archivo Excel
file_path = 'mef_ministers_dataset.xlsx'  # Asegrate de cambiar esta ruta
df = pd.read_excel(file_path)

# Asegrate de que las fechas estn en formato datetime
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')  # Convertir a datetime y manejar errores
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')

# Eliminar filas con fechas nulas
df = df.dropna(subset=['start_date', 'end_date'])

# Calcular la duracin en das
df['duration_days'] = (df['end_date'] - df['start_date']).dt.days

# Ordenar los ministros desde el ms antiguo hasta el ms reciente segn la fecha de inicio
df = df.sort_values('start_date', ascending=True)

# Definir la escala de colores personalizada
custom_colors = [
    "#66EFC4", "#00DFA2",  # Greens (recent)
    "#007e7d", "#005f5e",  # Dark greens
    "#FFECA8", "#FFF183",  # Yellows
    "#FFAAA0", "#ff8575",  # Pinks
    "#FF4C7F", "#E6004C"   # Reds (older)
]

# Crear un cmap de colores personalizados
custom_cmap = ListedColormap(custom_colors)

# Normalizacin de la duracin para el mapa de colores personalizado
norm = Normalize(vmin=300, vmax=900)

# Crear la figura para el grfico
fig, ax = plt.subplots(figsize=(14, 6))

# Graficar las lneas de tiempo para cada ministro, agrupando los periodos de presidentes
for i, row in df.iterrows():
    color = custom_cmap(norm(row['duration_days']))  # Usar el mapa de color personalizado para la duracin
    ax.plot([row['start_date'], row['end_date']], [i, i], linewidth=5, color=color)

# Ajustar el formato de las fechas en el eje X
ax.xaxis.set_major_locator(mdates.YearLocator(2))  # Mostrar cada 2 aos
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gcf().autofmt_xdate()

# Eje Y: Ministros, asegurndonos de que los ministros ms antiguos estn arriba
ax.set_yticks(np.arange(len(df)))
ax.set_yticklabels(df['minister_name'])

# Eje Y: Etiqueta
ax.set_ylabel("Ministro", fontsize=16)

# Eje X: No tiene etiqueta
ax.set_xlabel('')

# Rejilla esttica
ax.grid(True, color='#f5f5f5', alpha=0.75, linewidth=1)

# Lnea horizontal en 0 (simulando un tono neutral en el eje Y)
ax.axhline(0, color='#292929', linestyle='-', linewidth=1)

# Personalizacin de los ticks en el eje Y y X
ax.tick_params(axis='x', labelrotation=0, labelsize=12)  # Rotar etiquetas X a 0 (horizontal)
ax.tick_params(axis='y', labelsize=12)

# Leyenda personalizada (escala de colores de duracin en das)
sm = plt.cm.ScalarMappable(cmap=custom_cmap, norm=norm)
sm.set_array([])  # Necesario para la leyenda
cbar = fig.colorbar(sm, ax=ax, orientation='horizontal', pad=0.1)
cbar.set_label('Duracin en das', fontsize=12)
cbar.set_ticks([300, 600, 900])  # Los valores que aparecen en la leyenda de la escala

# Mover la leyenda a la parte superior izquierda
ax.legend(['Das en la Oficina'], loc='upper left', frameon=True)
ax.get_legend().get_frame().set_facecolor('white')
ax.get_legend().get_frame().set_alpha(0.5)
ax.get_legend().get_frame().set_edgecolor('#f5f5f5')  # contorno sin transparencia
ax.get_legend().get_frame().set_linewidth(2)  # grosor del contorno

# Ajuste final de la figura
plt.tight_layout()

# Guardar el grfico
plt.savefig('ministers_tenure.png', dpi=300)

# Mostrar el grfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import numpy as np
from matplotlib import cm
from matplotlib.colors import Normalize, LinearSegmentedColormap

# Cargar el archivo Excel
file_path = 'mef_ministers_dataset.xlsx'  # Asegrate de cambiar esta ruta
df = pd.read_excel(file_path)

# Asegrate de que las fechas estn en formato datetime
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')  # Convertir a datetime y manejar errores
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')

# Eliminar filas con fechas nulas
df = df.dropna(subset=['start_date', 'end_date'])

# Calcular la duracin en das
df['duration_days'] = (df['end_date'] - df['start_date']).dt.days

# Ordenar los ministros desde el ms antiguo hasta el ms reciente segn la fecha de inicio
df = df.sort_values('start_date', ascending=True)

# Crear un colormap de degradado con los colores especificados
colors = ["#033280", "#00DFA2", "#E6004C"]
cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)

# Normalizacin de la duracin para el mapa de colores
norm = Normalize(vmin=df['duration_days'].min(), vmax=df['duration_days'].max())

# Crear la figura para el grfico
fig, ax = plt.subplots(figsize=(14, 6))

# Graficar las lneas de tiempo para cada ministro
for i, row in df.iterrows():
    color = cmap(norm(row['duration_days']))  # Usar el mapa de color continuo para la duracin
    ax.plot([row['start_date'], row['end_date']], [i, i], linewidth=5, color=color)

# Ajustar el formato de las fechas en el eje X
ax.xaxis.set_major_locator(mdates.YearLocator(2))  # Mostrar cada 2 aos
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gcf().autofmt_xdate()

# Eje Y: Ministros, asegurndonos de que los ministros ms antiguos estn arriba
ax.set_yticks(np.arange(len(df)))  # 23 etiquetas por los 23 ministros
ax.set_yticklabels(df['minister_name'])

# Eje Y: Etiqueta
ax.set_ylabel("Ministro", fontsize=16)

# Eje X: No tiene etiqueta
ax.set_xlabel('')

# Rejilla esttica
ax.grid(True, color='#f5f5f5', alpha=0.75, linewidth=1)

# Lnea horizontal en 0 (simulando un tono neutral en el eje Y)
ax.axhline(0, color='#292929', linestyle='-', linewidth=1)

# Personalizacin de los ticks en el eje Y y X
ax.tick_params(axis='x', labelrotation=0, labelsize=12)  # Rotar etiquetas X a 0 (horizontal)
ax.tick_params(axis='y', labelsize=12)

# Leyenda personalizada (escala de colores de duracin en das)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Necesario para la leyenda
cbar = fig.colorbar(sm, ax=ax, orientation='horizontal', pad=0.1)
cbar.set_label('Duracin en das', fontsize=12)

# Ajustar los valores de la leyenda
cbar.set_ticks([df['duration_days'].min(), df['duration_days'].max()])
cbar.set_ticklabels([f"{int(df['duration_days'].min())}", f"{int(df['duration_days'].max())}"])

# Ajuste final de la figura
plt.tight_layout()

# Guardar el grfico
plt.savefig('ministers_tenure.png', dpi=300)

# Mostrar el grfico
plt.show()
```
